```python
apple_dates = apple['Date'].tolist()
tesla_dates = tesla['Date'].tolist()
print(apple_dates == tesla_dates)
```

```python
dates = apple_dates
```

```python
acceptable_dates = ['08/23/2017',
'02/23/2018',
'08/23/2018',
'02/22/2019',
'08/23/2019',
'02/24/2020',
'08/24/2020',
'02/24/2021',
'08/24/2021',
'02/22/2022',
'08/22/2022']
```

```python
r_a = []
r_t = []

for label, row in apple.iterrows():
    r_a.append(float(row['Close/Last'][2:]) / float(row['Open'][2:]))

for label, row in tesla.iterrows():
    r_t.append(float(row['Close/Last'][2:]) / float(row['Open'][2:]))
```

##### 1. Buy and Hold strategy with parameters (i/10 , (10-i)/10)

```python
init = 1000
final_wealth = []

for i in range(11):

    curr_wealth = []

    a_w = init * 0.1 * i
    t_w = init * 0.1 * (10-i)

    for j in range(len(dates)):
    
        a_w *= r_a[j]
        t_w *= r_t[j]

        if dates[j] in acceptable_dates:
            curr_wealth.append(a_w + t_w)

    final_wealth.append(curr_wealth)
```

```python
colors = ['b' , 'g' , 'r' , 'c' , 'm' , 'y' , 'k' , 'darkorange' , 'purple' , 'greenyellow' , 'slategrey']
for i in range(len(final_wealth)):
    plt.plot(acceptable_dates , final_wealth[i] , label = "i = {}".format(i) , color = colors[i])

plt.grid(True)
plt.legend()
plt.xticks(rotation = 45)
plt.title("Buy and Hold strategy with parameters (i/10 , (10-i)/10)")
plt.show()
```

##### 2. Cover's Universal Portfolio Algorithm

(a) $m+1$ CRPs with parameters (i/5 , (m-i)/5) for strategies $i = 0,1,2, \ldots , m$, and initial wealth is distriuted uniformly among all $m+1$ CRPs uniformly. Choose $m = 51, 101, 1001$.

```python
init = 1000
M = [51,101,1001]
final_wealth = []

for m in M:
    wealth = np.array([init/(m+1) for i in range(m+1)])
    b = np.array([[i/m , (m-i)/m] for i in range(m+1)]).T
    curr_wealth = []

    for i in range(len(dates)):

        for k in range(m+1):
            wealth[k] = wealth[k] * (r_a[i] * b[0][k] + r_t[i] * b[1][k])
        if dates[i] in acceptable_dates:
            curr_wealth.append(sum(wealth))

    final_wealth.append(curr_wealth)
```

```python

plt.plot(acceptable_dates , final_wealth[0] , 'b-' , label = "m = {}".format(M[0]) )
plt.plot(acceptable_dates , final_wealth[1] , 'g--' , label = "m = {}".format(M[1]) )
plt.plot(acceptable_dates , final_wealth[2] , 'r-.', label = "m = {}".format(M[2]) )


plt.grid(True)
plt.legend()
plt.xticks(rotation = 45)
plt.title("Cover's Universal Portfolio Algorithm with uniform distribution")
plt.show()
```

3. Repeat the above experiments with Dirichlet Distribution with parameters $(\frac{1}{2} , \frac{1}{2} , \ldots , \frac{1}{2})$ as initial distribution.

```python
init = 1000
M = [51,101,1001]
final_wealth = []

for m in M:
    b = np.array([[i/m , (m-i)/m] for i in range(m+1)]).T

    sz = 100000
    samples = np.random.dirichlet((1/2,1/2), size = sz)
    probs = np.zeros(m+1)
    for i in range(np.size(samples[:,1])):
        for j in range(m+1):
            if float((2*j+1)/(2*(m-1))) >= samples[:,1][i] and float((2*j-1)/(2*(m-1))) <= samples[:,1][i]:
                probs[j] = probs[j] + 1  
    probs = probs/sz

    wealth = np.array([init * probs[k] for k in range(m+1)])
    curr_wealth = []
    print(wealth)
    for i in range(len(dates)):

        for k in range(m+1):
            wealth[k] = wealth[k] * (r_a[i] * b[0][k] + r_t[i] * b[1][k])
        if dates[i] in acceptable_dates:
            curr_wealth.append(sum(wealth))

    final_wealth.append(curr_wealth)
```

```python
plt.plot(acceptable_dates , final_wealth[0] , 'b-' , label = "m = {}".format(M[0]) )
plt.plot(acceptable_dates , final_wealth[1] , 'g--' , label = "m = {}".format(M[1]) )
plt.plot(acceptable_dates , final_wealth[2] , 'r-.', label = "m = {}".format(M[2]) )


plt.grid(True)
plt.legend()
plt.xticks(rotation = 45)
plt.title("Cover's Universal Portfolio Algorithm with Dirichlet distribution")
plt.show()
```
