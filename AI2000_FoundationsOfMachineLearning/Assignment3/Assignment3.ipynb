Consider a neural network, as given in the figure below, using the sigmoid activation function, given by: $\sigma(x) = \frac{1}{1 + e^{-x}}$

![alt text](XOR_figure2.png)

Let the output  after hidden layer be given by $z$, and final output be given by $y$  and let the threshold $\theta = 0$ . Then, we can write the following equations.

$h_1 = w_{11}x_1 + w_{21}x_2 + b_1$

$h_2 = w_{12}x_1 + w_{22}x_2 + b_2$

$z = w_1'\sigma(w_{11}x_1 + w_{21}x_2 + b_1) + w_2'\sigma(w_{12}x_1 + w_{22}x_2 + b_2) + b' $

$(x_1 , x_2) = (1,1)$, $y = 0$ : $\implies w_1'\sigma(w_{11} + w_{21} + b_1) + w_2'\sigma(w_{12} + w_{22} + b_2) + b'< 0$

$w_1'= w_2'= -b'$

$w_{11} = w_{21} = w_1$

$w_1 + b_1 = 0$

$\sigma(0) + \sigma(0) \geq 1$

#### 2. Neural Networks
#### The extension of the cross entropy function for multi-class classification problem is given by:

$E(\vec{w}) = - \sum_{n=1}^N \sum_{k=1}^K t_{kn} ln y_k(\vec{x}_n , \vec{w})$

#### where $K$ is the number of classes, $N$ is the number of data samples, $\vec{t}_n$ is a one-hot vector which designates the expected output for data sample $\vec{x}_n$. The network outputs $y_k(\vec{x}_n , \vec{w}) = p(t_k = 1 | x)$ are given by the softmax activation function:

$y_k(\vec{x},\vec{w}) = \frac{exp(a_k(\vec{x}, \vec{w}))}{\sum_j exp(a_j(\vec{x}, \vec{w}))}$

#### satisfying $0 \leq y_k \leq 1$ and $\sum_k y_k = 1$ where $a_k$ are the pre- softmax activation functions of the output layer neurons. Show that the derivative of the above error function with respect to the activation $a_k$ for an output unit having a logistic sigmoid activation function is given by:

$\frac{\partial E}{\partial a_k} = y_k - t_k$

Answer:

Let us say the $n^{th}$ data point $\vec{x}_n$ belongs to class $j$ and let $t_{mn}$ represent the $m^{th}$ component of the one hot vector $\vec{t}_n$. Then, $\vec{t_{n}}$ would have $t_{jn} = 1$ and the rest of the components $ = 0 $. Thus, we can say $\sum_{m} t_{mn} = 1$. 

$E(\vec{w}) = -\sum_{n=1}^N \sum_{j=1}^K t_{jn} ln y_j(\vec{x}_n , \vec{w})$ 
 
$\implies\frac{\partial E(\vec{w})}{\partial a_k} =  -\sum_{n=1}^N \sum_{j=1}^K \frac{\partial}{\partial a_j}\left [t_{jn}ln y_{j}(\vec{x}_n , \vec{w}) \right]  = -\sum_{n=1}^N \sum_{j=1}^K \left[\frac{t_{jn}}{y_j(\vec{x}_n , \vec{w})} \times \frac{\partial}{\partial a_k} \left\{  y_j (\vec{x}_n , \vec{w} )\right\} \right]$

In general, the partial derivative of $y_l(\vec{x}, \vec{w})$ wrt $a_k$ can be calculated using quotient rule as:

$\frac{\partial y_l(\vec{x}, \vec{w})}{\partial a_k} = \frac{\partial}{\partial a_k} \left\{ \frac{exp(a_l(\vec{x}, \vec{w}))}{\sum_j exp(a_j(\vec{x}, \vec{w}))} \right\} = \frac{\frac{\partial}{\partial a_k}\left\{exp(a_l(\vec{x}, \vec{w})) \right\} \times \sum_j exp(a_j(\vec{x}, \vec{w})) - exp(a_l(\vec{x}, \vec{w})) \times exp(a_k(\vec{x}, \vec{w}))}{(\sum_j exp(a_j(\vec{x}, \vec{w})))^2}$

This leads to two cases:

Case 1: $l = k$ 

$\implies \frac{\partial y_l(\vec{x}, \vec{w})}{\partial a_k} = \frac{exp(a_l(\vec{x}, \vec{w})) \times \sum_j exp(a_j(\vec{x}, \vec{w})) - exp(a_l(\vec{x}, \vec{w})) \times exp(a_l(\vec{x}, \vec{w}))}{(\sum_j exp(a_j(\vec{x}, \vec{w})))^2} = \frac{exp(a_l(\vec{x}, \vec{w})) \times (\sum_j exp(a_j(\vec{x}, \vec{w})) - exp(a_l(\vec{x}, \vec{w})))}{(\sum_j exp(a_j(\vec{x}, \vec{w})))^2} $ 

$\implies \frac{\partial y_l(\vec{x}, \vec{w})}{\partial a_k} =\frac{exp(a_l(\vec{x} , \vec{w}))}{\sum_j exp(a_j(\vec{x}, \vec{w}))} \times \frac{\sum_j exp(a_j(\vec{x}, \vec{w})) - exp(a_l(\vec{x} , \vec{w}))}{\sum_j exp(a_j(\vec{x}, \vec{w}))} = y_l(\vec{x} , \vec{w})(1 - y_l(\vec{x},\vec{w}))$

Case 2: $l \neq k$

$\implies  \frac{\partial y_l(\vec{x}, \vec{w})}{\partial a_k} = \frac{- exp(a_l(\vec{x}, \vec{w})) \times exp(a_k(\vec{x}, \vec{w}))}{(\sum_j exp(a_j(\vec{x}, \vec{w})))^2} = \frac{- exp(a_l(\vec{x}, \vec{w}))}{\sum_j exp(a_j(\vec{x}, \vec{w}))} \times \frac{exp(a_k(\vec{x}, \vec{w}))}{\sum_j exp(a_j(\vec{x}, \vec{w}))} = -y_l(\vec{x}, \vec{w})y_k(\vec{x},\vec{w})$

Thus, the gradient of the loss function can now be written as:

$\frac{\partial E(\vec{w})}{\partial a_k} = -\sum_{n=1}^N \sum_{j=1}^K \left[\frac{t_{jn}}{y_j(\vec{x}_n , \vec{w})} \times \frac{\partial}{\partial a_k} \left\{  y_j (\vec{x}_n , \vec{w} )\right\} \right] = -\sum_{n=1}^N \left[ \sum_{j\neq k} \left[\frac{t_{jn}}{y_j(\vec{x}_n , \vec{w})} \frac{\partial y_j(\vec{x}_n , \vec{w})}{\partial a_k}\right] + \sum_{j=k}\frac{t_jn}{y_j(\vec{x}_n , \vec{w})} \frac{\partial y_j(\vec{x}_n , \vec{w})}{\partial a_k}\right]$

$\implies \frac{\partial E(\vec{w})}{\partial a_k} =  -\sum_{n=1}^N \left[ \sum_{j\neq k}\left[\frac{-t_{jn}}{y_j(\vec{x}_n , \vec{w})} y_j(\vec{x}_n, \vec{w})y_k(\vec{x}_n , \vec{w})\right] + \frac{t_{kn}}{y_k(\vec{x}_n , \vec{w})} y_k(\vec{x}_n, \vec{w})(1-y_k(\vec{x}_n , \vec{w}))\right]$

$\implies \frac{\partial E(\vec{w})}{\partial a_k} = \sum_{n=1}^N \left[ \sum_{j\neq k}\left[t_{jn}y_k(\vec{x}_n , \vec{w})\right] - t_{kn}(1 - y_k(\vec{x}_n , \vec{w}))\right]$

$\implies \frac{\partial E(\vec{w})}{\partial a_k} = \sum_{n=1}^N \left[ \sum_{j\neq k}\left[t_{jn}y_k(\vec{x}_n , \vec{w})\right] - t_{kn} + t_{jn}y_j(\vec{x}_n , \vec{w})\right]$ since for the right-most terms we had assumed $j = k$

$\implies \frac{\partial E(\vec{w})}{\partial a_k} = \sum_{n=1}^N \left[ \sum_{j=1}^K\left[t_{jn}y_k(\vec{x}_n , \vec{w})\right] - t_{kn}\right]$

$\implies \frac{\partial E(\vec{w})}{\partial a_k} = \sum_{n=1}^N \left[ y_k(\vec{x}_n , \vec{w})\sum_{j=1}^Kt_{jn} - t_{kn}\right]$

$\implies \frac{\partial E(\vec{w})}{\partial a_k} = \sum_{n=1}^N \left[ y_k(\vec{x}_n , \vec{w})- t_{kn}\right]$

$t_{kn}$ is the $k^{th}$ component of $\vec{t}_n$ for the $n^{th}$ datapoint $\vec{x}_n$. Thus, for a specific datapoint $\vec{x}_n$, the gradient of the loss function is given by:

$\frac{\partial E}{\partial a_k} = y_k - t_k$

Hence, proved.

#### 3. Ensemble Methods
#### Consider a convex function $f(x) = x^2$. Show that the average expected sum of squares error $E_{AV} = \frac{1}{M} \sum _{m=1}^M \mathbb{E}_x [(y_m(x) - f(x))^2] $ of the members of an ensemble method and the expected error $E_{ENS} =  \mathbb{E}_x[(\frac{1}{M}\sum _{m=1}^M  y_m(x) - f(x))^2]$ of the ensemble satisfy:
$E_{ENS} \leq E_{AV}$

Further, show that the following result holds for any error function $E(y)$ as long as it is convex.

Answer: 

Subsituting the values of $E_{AV}$ and $E_{ENS}$, and expanding the squares, we get:

$\mathbb{E}_x[\frac{1}{M^2}(\sum _{m=1}^M  y_m(x))^2 + f^2(x) - \frac{2}{M}f(x)\sum_{m=1}^My_m(x)] \leq \frac{1}{M} \sum _{m=1}^M \mathbb{E}_x [y_m^2(x) + f^2(x) - 2y_m(x)f(x)]$

Using Linearity of Expectation:

$\frac{1}{M^2}\mathbb{E}_x[(\sum _{m=1}^M  y_m(x))^2] + \mathbb{E}_x[f^2(x)] -  \frac{2}{M}\mathbb{E}_x[f(x)\sum_{m=1}^My_m(x)] \leq \frac{1}{M}\sum_{m=1}^M\mathbb{E}_x[y_m^2(x)] +  \frac{1}{M}\sum_{m=1}^M\mathbb{E}_x[f^2(x)] - \frac{2}{M}\sum_{m=1}^M\mathbb{E}_x[y_m(x)f(x)]$

Since, $\mathbb{E}_x[f^2(x)]$ is independent of $m$ , $ \frac{1}{M}\sum_{m=1}^M\mathbb{E}_x[f^2(x)] = \frac{1}{M}\mathbb{E}_x[f^2(x)]\sum_{m=1}^M1 = \mathbb{E}_x[f^2(x)]$, which then cancels the term on the left side of the inequality. We get:

$\frac{1}{M^2}\mathbb{E}_x[(\sum _{m=1}^M  y_m(x))^2] -  \frac{2}{M}\mathbb{E}_x[f(x)\sum_{m=1}^My_m(x)] \leq \frac{1}{M}\sum_{m=1}^M\mathbb{E}_x[y_m^2(x)] - \frac{2}{M}\sum_{m=1}^M\mathbb{E}_x[y_m(x)f(x)]$

Again, using linearity of expectation, we can show $\frac{2}{M}\mathbb{E}_x[f(x)\sum_{m=1}^My_m(x)] = \frac{2}{M}\sum_{m=1}^M\mathbb{E}_x[f(x)y_m(x)]$, cancelling the term on the right side of the inequality. We finally get:

$\frac{1}{M^2}\mathbb{E}_x[(\sum _{m=1}^M  y_m(x))^2] \leq \frac{1}{M}\sum_{m=1}^M\mathbb{E}_x[y_m^2(x)]$

To prove this inequality, we will use Jensen's Inequality
which states that for any convex function $g(x)$ and for non-negative numbers $p_i$ such that $\sum_{i=1}^n p_i = 1$:

$g(\sum_{i=1}^n p_ix_i) \leq \sum_{i=1}^n p_ig(x_i)$

We can rewrite the statement to prove as:

$\mathbb{E}_x[(\sum _{m=1}^M \frac{1}{M}  y_m(x))^2] \leq  \sum _{m=1}^M \frac{1}{M}\mathbb{E}_x [(y_m^2(x))]$

On comparing with the statement to prove, we find that $g(x) = \mathbb{E}_x(x^2)$ and $p_i = \frac{1}{M}$, where $\sum_{m=1}^M \frac{1}{M} = 1$. So, if we can prove $\mathbb{E}_x(x^2)$ is convex, we can prove the statement in general:

We know, $x^2$ is convex because $\frac{d^2 }{dx^2} (x^2) = 2 > 0 \forall x$.

Let us say $\mathbb{E}_x[x^2] = \sum_{i=1}^n u_ix_i^2$ for some non- negative $u_i$. Then,

$\frac{d^2}{dx^2}(\mathbb{E}_x[x^2]) = \sum_{i=1}^n 2u_i > 0 \forall x$.

Thus, $\mathbb{E}_x(x^2)$ is convex, and using Jensen's Inequality, we can arrive at the conclusion that:

$E_{ENS} \leq E_{AV}$



Hence, proved.

<u>To prove: </u> $\mathbb{E}_x[E(\frac{1}{M}\sum _{m=1}^M  y_m(x))] \leq \frac{1}{M} \sum _{m=1}^M \mathbb{E}_x [E(y_m(x))]$ where $E(y)$ is the error function, and is convex in $y$.

<u>Proof :</u> Using Jensen's Inequality, we know that for any convex function $g(x)$ and for non-negative numbers $p_i$ such that $\sum_{i=1}^n p_i = 1$:

$g(\sum_{i=1}^n p_ix_i) \leq \sum_{i=1}^n p_ig(x_i)$

We can rewrite the statement to prove as:

$\mathbb{E}_x[E(\sum _{m=1}^M \frac{1}{M}  y_m(x))] \leq  \sum _{m=1}^M \frac{1}{M}\mathbb{E}_x [E(y_m(x))]$

On comparing with the statement to prove, we find that $g(x) = \mathbb{E}_x(E(x))$ and $p_i = \frac{1}{M}$, where $\sum_{m=1}^M \frac{1}{M} = 1$. So, if we can prove $\mathbb{E}_x(E(x))$ is convex, we can prove the statement in general:

Let us say $\mathbb{E}_x[E(x)] = \sum_{i=1}^n u_iE_i(x)$ for some non- negative $u_i$. Then,


Since $E(x)$ is convex, we know $\frac{d^2E}{dx^2} \geq 0 \forall x$

Now, $\mathbb{E}_x(E(x)) = \sum u_iE_i(x) \implies \frac{d^2\mathbb{E}_x(E)}{dx^2} = \sum u_i\frac{d^2E_i}{dx^2} \geq 0 \forall x $

Thus, $g(x) = \mathbb{E}_x(E(x))$ is convex, and using Jensen's Inequality, we can say:

$g(\sum_{i=1}^n p_ix_i) \leq \sum_{i=1}^n p_ig(x_i) \implies \mathbb{E}_x[E(\frac{1}{M}\sum _{m=1}^M  y_m(x))] \leq  \sum _{m=1}^M \frac{1}{M}\mathbb{E}_x [E(y_m(x))]$

Hence, proved.

```python
# reading the data
df = pd.read_table("https://web.stanford.edu/~hastie/ElemStatLearn//datasets/spam.data", \
delimiter = " ", header = None)

# splitting the data into training and testing data
train , test = sklearn.model_selection.train_test_split(df , train_size = 0.7 , test_size = 0.3 \
    , random_state = 100)

# converting to numpy arrays
train = train.to_numpy()
test = test.to_numpy()

# splitting the data into features and labels
X_train = train[: , :-1]
y_train = train[: , -1]
X_test = test[: , :-1]
y_test = test[: , -1]

```

```python
# defining the class

MAX_DEPTH = 5

class Node:
    def __init__(self,depth):
        self.is_terminal_node = False
        self.left = None
        self.right = None
        self.split_index = None
        self.split_threshold = None
        self.label = 0
        self.depth = depth
        pass

    def train(self , X,y,m):
        # self.label = majority label of training data
        
        ones = np.count_nonzero(y==1)
        zeros = len(y) - ones
        if ones > zeros:
            self.label = 1
        else:
            self.label = 0 
        # If max depth reached, just stop ->is_terminal_node = True
        if self.depth == MAX_DEPTH:
            self.is_terminal_node = True
        # Calculate initial entropy
        entropy_init = entropy(y)
        
        # If initial entropy>0.1 -> Find your best split -> splitindex,threshold 
        if entropy_init>0.1 and self.is_terminal_node == False:
            entropy_max = 100
            # looping over all attributes
            X_yes_final = []
            y_yes_final = []
            X_no_final  = []
            y_no_final = []

            random_columns = np.random.choice(X.shape[1] , size = m , replace = False)
            for column in random_columns:
                maxi = np.max(X[: , column])
                mini = np.min(X[: , column])    
                potential_thresholds = np.linspace(mini, maxi , 20)
                # looping over all thresholds
                for t in potential_thresholds:
                    yes = np.where(X[: , column]>=t)[0] # greater than threshold
                    no = np.where(X[: , column]<t)[0]   # lesser than threshold
                    if len(yes)==0 or len(no)==0:
                        continue
                    
                    y_yes = np.array([y[i] for i in yes])
                    y_no = np.array([y[i] for i in no])
                    total = len(y_yes) + len(y_no)
                    entropy_yes = entropy(y_yes)
                    entropy_no = entropy(y_no)
                    
                    entropy_final = (len(y_yes)/total)*entropy_yes + (len(y_no)/total)*entropy_no
                    # maximising information gain -> minimising entropy
                    if entropy_final<entropy_max:
                            self.split_index = column
                            self.split_threshold = t
                            if ones>zeros:
                                self.label = 1
                            else:
                                self.label = 0
                            entropy_max = entropy_final
                            X_yes_final = np.array([X[i] for i in yes])
                            X_no_final = np.array([X[i] for i in no] )  
                            y_yes_final = np.array([y[i] for i in yes])
                            y_no_final = np.array([y[i] for i in no])
                            #print("label = ",self.label)
        
            # assuming left path corresponds to data greater than threshold
        
            if(len(y_yes_final) > 0): 
                l = Node(self.depth+1)
                self.left = l
                l.train(X_yes_final , y_yes_final, m)
            if   (len(y_no_final)>0):
                r = Node(self.depth+1)
                self.right = r
                r.train(X_no_final , y_no_final, m)
            
        
        else:
            self.is_terminal_node = True

    # implement this function
    def classify(self, test_instance):
        result = 0 # baseline: always classifies as 0
        
        if(self.is_terminal_node==True):
            #print("label = ", self.label)
            result = self.label
        if self.split_threshold is not None and self.is_terminal_node==False :
            if test_instance[self.split_index]>=self.split_threshold:
                return self.left.classify(test_instance)
            else:
                return self.right.classify(test_instance)

        return result



def entropy(y):
    ones = np.count_nonzero(y==1)/len(y)

    zeros = 1 - ones
    if ones==0 or zeros==0:
        entropy = 0
    else:
        entropy = ones*math.log2(ones) + zeros * math.log2(zeros)
    return -entropy
```

#### (a)- random forest

We will use a forest of 100 trees with m = 10

```python
# part (a)

m = 10
# to keep track of all trees
forest_trees = []

# starting time
start = time.time()

# training and storing all trees:
for i in range(100):
    # sampling the training data
    sampled_rows = np.random.choice(a = X_train.shape[0] , size = X_train.shape[0] , replace = True)
    sampled_X_train = np.array([X_train[j] for j in sampled_rows])
    sampled_y_train = np.array([y_train[j] for j in sampled_rows])
    tree = Node(0)
    tree.train(sampled_X_train, sampled_y_train,m)
    forest_trees.append(tree)

final_results = []
# testing over all trees
for instance in X_test:
    tree_results = []
    for j in range(100):
        tree_results.append(forest_trees[j].classify(instance))
    tree_results = np.asarray(tree_results)
    spam = np.count_nonzero(tree_results==1)
    non_spam = len(tree_results)-spam
    if spam>non_spam:
        final_results.append(1)
    else:
        final_results.append(0)

# ending time
end = time.time()

# predicting accuracy of our model:
accuracy = 0
for i in range(len(y_test)):
    if y_test[i]==final_results[i]:
        accuracy+=1
accuracy = accuracy/len(y_test)*100

print("*"*75)
print("TREE BUILT BY AUTHOR:")
print("Accuracy: %.4f" % accuracy)
print("Time Elapsed = ", end-start)
print("*"*75)
```

#### 5. Gradient Boosting

```python
# reading the training and testing data

train_data = pd.read_csv("loan_train.csv")
print(train_data.shape)
test_data = pd.read_csv("loan_test.csv")
print(test_data.shape)
```

*****************************************************************
#### (a)- Preprocessing the data:
Answer: 

The steps I have taken in order are:
1. Remove all rows having loan_status as "Current".

2. Convert loan_status column to binary features, i.e if loan_status is "Fully Paid", then it represents class $+1$, else it represents class $-1$.

3. Removed all columns having datatype as "object".

4. Removed columns having all values as NaN.

5. Removed columns having more than 5000 NaN values, since these columns would not be contributing significantly to the predictions.

6. If any NaNs remain, they have been replaced with the mean of their respective columns.

7. For the test dataset, we keep only those columns remaining in the training dataset.

8. Finally, if there are any NaNs in the test dataset, they have been replaced with the mean of their respective columns.
*****************************************************************

```python
# part(a)


# dropping rows with loan_status as Current
train_data.drop(train_data[train_data['loan_status'] == "Current"].index, inplace = True)
test_data.drop(test_data[test_data['loan_status'] == "Current"].index, inplace = True)

# convertng loan status column to binary features
train_data['loan_status'] = train_data['loan_status'].replace(["Fully Paid", "Charged Off"] , [+1 , -1])
test_data['loan_status'] = test_data['loan_status'].replace(["Fully Paid", "Charged Off"] , [+1 , -1])
#print(train_data.shape)

# removing all columns with datatype as object
for column in train_data:
    if train_data[column].dtypes == object:
        train_data.drop(column , axis = "columns" , inplace = True)

# removing all columns with all Nans
train_data.dropna(axis= "columns", how = "all", inplace = True)

# removing all columns with more than 5000 NaNs
columns_with_Nans = train_data.isna().sum()>50

# we now drop those columns from train_df
for column in train_data.columns:
    if columns_with_Nans[column] == True:
        train_data.drop(column, axis=1 , inplace=True)

# filling entries still having Nans with mean of that column
train_data.fillna(train_data.mean(), inplace = True)

# keeping only those columns in test_data, which are there in train_data
test_data = test_data[train_data.columns]

# filling any remaining Nans in test_data with mean of column
test_data.fillna(test_data.mean(), inplace = True)

columns_without_target = [column for column in train_data.columns if column != "loan_status"]
X_train = np.array(train_data[columns_without_target])
y_train = np.array(train_data["loan_status"])
X_test = np.array(test_data[columns_without_target])
y_test = np.array(test_data["loan_status"])
```

#### (b) Gradient boosting

```python
# part (b)
precision = []
recall = []
test_accuracy = []


print ("*"*75)
number_of_trees = [100* i for i in range(1,10)]

for n_trees in number_of_trees:
    clf = sklearn.ensemble.GradientBoostingClassifier(n_estimators=n_trees)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    #calculating test accuracy
    t_a = np.sum(y_pred==y_test)/len(y_test)
    #t_a*=100
    test_accuracy.append(t_a)

    #calculating precision and recall
    TP = FP = FN = 0
    for i in range(len(y_test)):
        if y_pred[i] == 1 and y_test[i]==1:
            TP+=1
        if y_pred[i]==1 and y_test[i]==-1:
            FP+=1
        if y_pred[i]==-1 and y_test[i] ==1:
            FN +=1
    p = TP/(TP+FP)
    r = TP/(TP+FN)
    print("For {} trees, the precision is {}".format(n_trees , p))
    print("For {} trees, the recall is {}".format(n_trees , r))
    precision.append(p)
    recall.append(r)

    index = np.where(test_accuracy == np.max(test_accuracy))[0][0]

print("THE BEST PERFORMANCE WAS REPORTED FOR {} NUMBER OF TREES, RESULTING IN A TEST \
    ACCURACY OF {}".format(number_of_trees[index], test_accuracy[index]))
print("*"*75)
```

```python
# plotting graphs for precision, recall and test_accuracies vs number of trees

plt.plot(number_of_trees , precision, 'r-', label = "precision")
plt.plot(number_of_trees , recall, 'b-', label = "recall")
plt.grid(True)
plt.legend()
plt.title("Precision and Recall vs Number of Estimators")
plt.show()

plt.plot(number_of_trees , test_accuracy, 'r-')
plt.grid(True)
plt.title("Test Accuracies vs Number of Estimators")
plt.show()
```
