#### Q2. PCA

#### (a) Let $\vec{X}'= [\vec{X_1} , \vec{X_2}, \ldots , \vec{X_p}]$ have covariance matrix $\sum$, with eigenvalue-eigenvectir pairs $(\lambda_1 , e_1) , (\lambda_2 , e_2) , \ldots ,(\lambda_p ,e_p)$ where $\lambda_1 \geq \lambda_2 \geq \ldots \lambda_p \geq 0$

#### Let $\vec{Y_1} = \vec{e_1}'\vec{X}, \vec{Y_2} = \vec{e_2}'\vec{X}, \ldots ,\vec{Y_p} = \vec{e_p}'\vec{X}$ be the principal components. Prove that:

$\sigma_{11} + \sigma_{22} + \ldots \sigma_{pp} = \sum_{i=1}^ p Var(\vec{X_i}) = \lambda_1 + \lambda_2 + \ldots + \lambda_p = \sum_{i=1}^p Var(\vec{Y_i})$

Answer: 

We know, the sum of diagonal entries of a matrix is equal to the sum of the eigenvalues of a matrix. Thus, we get:

$\sigma_{11} + \sigma_{22} + \ldots \sigma_{pp} = \lambda_1 + \lambda_2 + \ldots + \lambda_p$

Now, 

$Var(\vec{Y_i}) = \mathbb{E}[(\vec{Y_i} - \mathbb{E}[\vec{Y_i}])^2]$

$Var(\vec{Y_i}) = \mathbb{E}[(\vec{e_i'}\vec{X} - \mathbb{E}[\vec{e_i}\vec{X}])^2] = \mathbb{E}[(\vec{e_i'}\vec{X})^2]$ since mean of data is zero.

$Var(\vec{Y_i}) = \mathbb{E}[(\vec{e_i'}\vec{X})(\vec{e_i'}\vec{X})'] = \mathbb{E}[\vec{e_i'}\vec{X}\vec{X}'\vec{e_i}] = \mathbb{E}[\vec{e_i'}\sum\vec{e_i}]$ since $\vec{X}\vec{X}' = \sum$ is the covariance matrix

$Var(\vec{Y_i}) = \mathbb{E}[\vec{e_i'}\lambda_i\vec{e_i}]$ since $\sum \vec{e_i} = \lambda_i \vec{e_i}$

$Var(\vec{Y_i}) = \lambda_i\mathbb{E}[\vec{e_i'}\vec{e_i}] = \lambda_i$ since all eigenvectors are unit norm.

$\implies \sum_{i=1}^p Var(\vec{Y_i}) = \sum_{i=1}^p \lambda_i = \lambda_1 + \lambda_2 + \ldots + \lambda_p $

Finally, $Var(\vec{X_i})$ is the $i^{th}$ diagonal entry of the covariance matrix $\sum$, i.e 

$Var(\vec{X_i}) = \sigma_{ii} \implies \sum_{i=1}^p Var(\vec{X_i}) = \sum_{i=1}^p \sigma_{ii} = \sigma_{11} + \sigma_{22} + \ldots + \sigma_{pp} $

Thus, we have shown that all the 4 terms are equal.

#### (b) Suppose the random variables $\vec{X_1} , \vec{X_2} , \vec{X_3}$ have covariance matrix:

$\sum  = \left[\begin{matrix}
1 & -2 & 0\\
-2 & 5 & 0\\
0 & 0 & 2
\end{matrix}\right]$ 

#### The eigenvalues-eigenvector pairs are:

$\lambda_1 = 5.83;$ $ \vec{e_1}'= [0.383 , -0.924 , 0]$

$\lambda_2 = 2.0; $ $\vec{e_2}'= [0,0,1]$

$\lambda_3 = 0.17; $ $\vec{e_3}'= [0.924 , 0.383 , 0]$

#### (i) Find out the principal components $\vec{Y_1} , \vec{Y_2} , \vec{Y_3}$.

Answer:

$\vec{Y_1} = \vec{e_1}^T\vec{X}$

$\vec{Y_2} = \vec{e_1}^T\vec{X}$

$\vec{Y_3} = \vec{e_1}^T\vec{X}$

where $\vec{X}^T = [\vec{X_1} \vec{X_2} \vec{X_3}]$

#### (ii) Do you think $\vec{X_3}$ is a principal component?

Answer: 

Yes, $\vec{X_3}$ is a principal component, as evident by $\vec{e_2'}$. This can be seen because the co-variance of $\vec{X_3}$ wrt $\vec{X_1}$ and $\vec{X_2}$ is zero, and is thus, independent of $\vec{X_1}$ and $\vec{X_2}$.

#### (iii) Demonstrate $Var(\vec{Y_i}) = \lambda_i$, $i = 1,2,3$ and $Cov(\vec{Y_i} , \vec{Y_k}) = 0,$ $i \neq k$

Answer: 

In PCA, we know that every principal component is orthogonal to one another to capture maximum information that is uncorelated to the other principal components. This can also be seen by the fact that the covariance matrix is symmetric, and a symmetric matrix has orthogonal eigenvectors. Therefore, 

$\vec{e_i}^T\vec{e_k}=0$, $i \neq k$

$Cov(\vec{Y_i} , \vec{Y_k}) = Cov(\vec{e_i}^T\vec{X} , \vec{e_k}^T\vec{X}) $

$\implies Cov(\vec{Y_i} , \vec{Y_k}) = \mathbb{E}[(\vec{e_i}^T\vec{X} - \mathbb{E}[\vec{e_i}^T\vec{X}])(\vec{e_k}^T\vec{X} - \mathbb{E}[\vec{e_k}^T\vec{X}])^T] = \mathbb{E}[(\vec{e_i}^T\vec{X})(\vec{e_k}^T\vec{X})^T]$ since mean of data is assumed to be 0.

$\implies Cov(\vec{Y_i} , \vec{Y_k}) = \mathbb{E}[\vec{e_i}^T\vec{X}\vec{X}^T\vec{e_k}] = \mathbb{E}[\vec{e_i}^T\sum\vec{e_k}]$ where $\vec{X}\vec{X}^T = \sum$ is the covariance matrix

$\implies Cov(\vec{Y_i} , \vec{Y_k}) = \mathbb{E}[\lambda_k\vec{e_i}^T\vec{e_k}]$ since $\sum\vec{e_k} = \lambda_k\vec{e_k}$

$\implies Cov(\vec{Y_i} , \vec{Y_k}) = \begin{cases}
0 & i \neq k\\
\mathbb{E}[\lambda_i] = \lambda_i & i=k
\end{cases}$ since the eigenvectors are unit norm.

Also, $ Cov(\vec{Y_i} , \vec{Y_k})$ when $i = k$ is equal to $Var(\vec{Y_i})$.

Thus, $Var(\vec{Y_i}) = \lambda_i$, $i = 1,2,3$ and $Cov(\vec{Y_i} , \vec{Y_k}) = 0,$ $i \neq k$

#### (iv) Do you think any of the principal components can be ignored?

Answer: 

Yes, $\vec{Y_3}$ can be ignored. Since we know, each principal component $\vec{Y_i}$ contributes $\lambda_i$ of the variance, we find that the first two principal components capture the maximum information as shown:

Information captured by first two first principal components = 

$\frac{\lambda_1 + \lambda_2}{\lambda_1 + \lambda_2 + \lambda_3} = \frac{7.83}{8} = 0.97875$

Thus, the first two components capture 97.875% of the information, and we lose only 2.125% of the information, which is pretty decent. Thus, we can ignore the $\vec{Y_3}$ component.

Answer:

Since $y^{(pr)} \sim \mathcal(\mu_p , \sigma_p^2) \implies \mathbb{E}[y^{(pr)}] = \mu_p , Var[y^{(pr)}] = \sigma_p^2$ 

$z^{(pr)} \sim \mathcal(\nu_r , \tau_r^2) \implies \mathbb{E}[z^{(pr)}] = \nu_r , Var[z^{(pr)}] = \tau_r^2$ 

Consider $x^{(pr)} \sim \mathcal{N}(y^{(pr)} + z^{(pr)} , \sigma^2) \implies x^{(pr)} \sim \mathcal{N}(0 , \sigma^2) + y^{(pr)} + z^{(pr)}$

$\implies x^{(pr)} = \epsilon + y^{(pr)} + z^{(pr)}$ where $\epsilon \sim \mathcal{N}(0 , \sigma^2)$

$\implies \mathbb{E}[x^{(pr)}] = \mathbb{E}[\epsilon] + \mathbb{E}[y^{(pr)}] + \mathbb{E}[z^{(pr)}] = 0 + \mu_p + \nu_r = \mu_p + \nu_r$

$\implies Var[x^{(pr)}] =Var[\epsilon] + Var[y^{(pr)}] + Var[z^{(pr)}] = \sigma^2 + \sigma_p^2 + \tau_r^2 $

$\mu = \left[\begin{matrix}
\mu_p\\
\nu_r\\
\mu_p + \nu_r
\end{matrix}\right]$

$Cov[y^{(pr)} , z^{(pr)}] = 0$

$Cov[x^{(pr)} , z^{(pr)}] = Cov[\epsilon + y^{(pr)} + z^{(pr)} , z^{(pr)}]$

$Cov[x^{(pr)} , z^{(pr)}] = Cov[\epsilon , z^{(pr)}] + Cov[y^{(pr)},z^{(pr)}] + Cov[z^{(pr)},z^{(pr)}] = 0 + 0 + \tau^2_r = \tau_r^2$

$Cov[x^{(pr)} , y^{(pr)}] = Cov[\epsilon + y^{(pr)} + z^{(pr)} , y^{(pr)}]$

$Cov[x^{(pr)} , y^{(pr)}] = Cov[\epsilon , y^{(pr)}] + Cov[y^{(pr)},y^{(pr)}] + Cov[z^{(pr)},y^{(pr)}] = 0 + \sigma_p^2 + 0 = \sigma_p^2$

$\sum = \left[\begin{matrix}
\sigma_p^2 & 0 & \sigma_p^2\\
0 & \tau_r^2 & \tau_r^2\\
\sigma_p^2 & \tau_r^2 & \sigma_p^2 + \tau_r^2 + \sigma^2
\end{matrix}\right]$

#### (b) Implementing DBSCAN algorithm

```python
def distance(i , j):
    return math.sqrt((i[0]-j[0])**2 + (i[1]-j[1])**2)
```

```python
def DBSCAN(data,minPoints,eps):
    
    # defining arrays to keep track of visited points, neighbour points and clusters
    visited = (np.zeros(data.shape[0] , dtype = int))
    type =  (np.zeros(data.shape[0] , dtype = int))
    neighbour_points = []
    clusters = (np.zeros(data.shape[0] , dtype = int) - np.ones(data.shape[0] , dtype = int))
        # to distinguish if a point has been assigned a cluster or not we have a negative number
    
    
    # making a list of all neighbours
    for i in range(len(data)):     
        neighbours = []

        for j in range(len(data)):
            if distance(data[i] , data[j]) <= eps and i != j :
                neighbours.append(j)

        neighbour_points.append(neighbours)

    # we use the following convention:
    # type point = 0 implies point is a noise point
    # type point = 1 implies point is a border point
    # type point = 2 implies point is a core point

    # we now distinguish between core, border and noise points

    # we start with finding all core points
    for i in range(len(data)):
        if len(neighbour_points[i]) >= minPoints:
            type[i] = 2

    # now that we know core points, we can find border points
    for i in range(len(data)):
        if type[i] != 2: # not a core point
            for j in neighbour_points[i]:
                if type[j] == 2: # one of the neighbours is a core point
                    type[i] = 1
                    


    # we begin assigning the clusters
    cluster_number = -1
    for i in range(len(data)):
       
    
        if type[i] == 2 and visited[i] == 0: # unvisited core point
            visited[i] = 1
            current_cluster = -1
            for j in neighbour_points[i]:
                # we check if the point is not in the vicinity of another point 
                # that has been visited
                if visited[j] == 1:
                    current_cluster = clusters[j]
            
            # if still current_cluster is -1, we start with a new cluster
            if current_cluster == -1:
                cluster_number += 1
                current_cluster = cluster_number
                

            for j in neighbour_points[i]:
                clusters[j] = current_cluster
                visited[j] = 1
        
        elif type[i] == 2 and visited[i] == 1:
            current_cluster = clusters[i]
            for j in neighbour_points[i]:
                clusters[j] = current_cluster
                visited[j] = 1


        elif type[i] == 1 and visited[i] == 1: # visited border point
            current_cluster = clusters[i]
            for j in neighbour_points[i]:
                clusters[j] = current_cluster
                visited[j] = 1
        
        elif type[i] == 0 and visited[i] == 0: # unvisited noise point
            visited[i] = 1
            current_cluster = -1
            for j in neighbour_points[i]:
                # we check if the point is not in the vicinity of another point 
                # that has been visited
                if visited[j] == 1:
                    current_cluster = clusters[j]
            
            # if still current_cluster is -1, we start with a new cluster
            if current_cluster == -1:
                cluster_number += 1
                current_cluster = cluster_number
                

            for j in neighbour_points[i]:
                clusters[j] = current_cluster
                visited[j] = 1

        elif type[i] == 0 and visited[i] == 1: # visited noise point
            current_cluster = clusters[i]
            for j in neighbour_points[i]:
                clusters[j] = current_cluster
                visited[j] = 1
    
        # we have nothing to do with unvisited border points, 
        # since they will eventually get allotted as they are in the vicinity of some core point

    return clusters
```

#### (c) What are the differences between KMeans and DBSCAN?

Answer:

Clearly, there are two seperate clusters of points which KMeans is not able to take into account for. KMeans is not able to take the curvature of the shape of points into account, whereas DBSCAN is able to. KMeans takes points from both clusters close enough to one another, and puts them in one cluster, because it assumes a centroid, and based on the distances from that centroid, it assigns the cluster. It is very difficult to bring in the curvature while calculating distance form a point. On the other hand, DBSCAN is dependent on the density of points found in an epsilon - neighbourhood, and it can thus, cluster effieciently, when the data has arbitrary shape.

#### (b) Explain why different runs of t-SNE algorithm with the same hyperparametres produce different results.

Answer: 

This is because the cost function is a non-convex function. Whenever there are different initialisations, there is no guarantee that we would reach the global optimum. We may end up at a local optimum, saddle point or even a region that is flat but has non-zero gradient. For the best results, we should start from near the global optimum that guarantees convergence to the global optimum.
